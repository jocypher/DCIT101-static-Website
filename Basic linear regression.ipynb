```python
import matplotlib.pyplot as plt
import numpy as np
x = [1, 33, 60, 4.6]
y = [6, 82, 100, 19]
plt.plot(x, y, 'ro')
plt.axis([0, 100, 0, 120])
plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique (x)))
plt.show()
```
